```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Admin\Desktop\My stuff\Programming\ML\Projects\spotify-recommendation-system\data\processed\tracks_clean.csv")
```

```python
#Convert duration from milliseconds to minutes
df['duration_min']=df['duration_ms']/60000
```

```python
#Scale the audio features
from sklearn.preprocessing import MinMaxScaler

feature_cols=['danceability', 'energy', 'valence', 'acousticness',
                'tempo', 'loudness', 'speechiness', 'instrumentalness',
                'liveness', 'popularity']

scaler= MinMaxScaler()
df_scaled= df.copy()
df_scaled[feature_cols]= scaler.fit_transform(df[feature_cols])
```

```python
#Save the scaler
import pickle

with open(r"C:\Users\Admin\Desktop\My stuff\Programming\ML\Projects\spotify-recommendation-system\models\scaler.pkl", 'wb') as f:
    pickle.dump(scaler,f)
```

```python
df_scaled.to_csv(r"C:\Users\Admin\Desktop\My stuff\Programming\ML\Projects\spotify-recommendation-system\data\processed\tracks_feature.csv", index=False)
```
